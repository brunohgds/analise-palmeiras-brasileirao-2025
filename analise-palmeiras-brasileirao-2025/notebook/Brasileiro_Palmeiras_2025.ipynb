# Análise de Desempenho do Palmeiras no Campeonato Brasileiro 2025

## 1. Definição do Problema de Negócio

### 1.1. Contexto

O Campeonato Brasileiro de Futebol é uma das ligas mais competitivas do mundo, e o desempenho de um clube ao longo da temporada é de grande interesse para torcedores, analistas esportivos e a própria gestão do time. A capacidade de analisar dados de partidas pode revelar padrões, pontos fortes e fracos, e auxiliar na tomada de decisões estratégicas.

Este projeto foca na análise do desempenho do Sociedade Esportiva Palmeiras no Campeonato Brasileiro de 2025, utilizando dados públicos de resultados de jogos. O objetivo é transformar esses dados brutos em *insights* claros sobre a performance do time.

### 1.2. Problema de Negócio

Para torcedores e a comissão técnica, entender o desempenho de um time vai além de apenas ver a tabela de classificação. Questões como:

- **Desempenho Geral:** Qual foi o balanço de vitórias, empates e derrotas do Palmeiras?
- **Performance em Casa vs. Fora:** O time joga melhor em casa ou como visitante? Há uma diferença significativa na pontuação ou no saldo de gols?
- **Eficiência Ofensiva e Defensiva:** Quantos gols o time marcou e sofreu? Qual a média de gols por partida?
- **Consistência ao Longo do Tempo:** Existem períodos (meses ou dias da semana) em que o desempenho é notavelmente melhor ou pior?

Responder a essas perguntas de forma estruturada e visual é fundamental para avaliar a campanha do time e identificar áreas para melhoria.

### 1.3. Objetivos do Projeto

O objetivo principal deste projeto é realizar uma Análise Exploratória de Dados (EDA) do desempenho do Palmeiras no Campeonato Brasileiro de 2025 para extrair *insights* relevantes.

Os objetivos específicos são:

- **Calcular Métricas de Desempenho:** Determinar o número total de vitórias, empates, derrotas e pontos conquistados pelo Palmeiras.
- **Analisar Gols:** Calcular o total de gols marcados (pró) e sofridos (contra), e o saldo de gols.
- **Comparar Local de Jogo:** Avaliar o desempenho do time em partidas jogadas em casa e fora de casa.
- **Identificar Tendências Temporais:** Analisar a performance do time por mês e por dia da semana para identificar padrões.
- **Identificar Vítimas e Carrascos:** Determinar quais times o Palmeiras mais venceu e mais perdeu.

### 1.4. Critérios de Sucesso

O projeto será considerado um sucesso ao entregar:

- Um *dataset* pré-processado com as informações de desempenho do Palmeiras.
- Métricas claras e concisas sobre o desempenho geral do time.
- Visualizações que comparem o desempenho em casa e fora, e ao longo do tempo.
- Identificação dos principais adversários (vítimas e carrascos).
- Conclusões baseadas em dados que respondam às perguntas de negócio iniciais.
- Dados prontos para importação em ferramentas de BI como o Power BI.

## 4. Pré-processamento e Engenharia de Atributos

### 4.1. Conversão de Tipos e Extração de Informações Temporais
A coluna `data` será convertida para o tipo datetime para permitir a extração de informações como mês e dia da semana, que podem ser úteis para identificar padrões.

```python
# Convertendo coluna 'data' para datetime
df_brasileirao_2025['data']= pd.to_datetime(df_brasileirao_2025['data'])
```

```python
# Criar a condição booleana para verificar a existencia de jogos não finalizados
condicao_nao_finalizados = df_brasileirao_2025['status'] != "FINISHED"
print(f"Quantidade de jogos não finalizados: {condicao_nao_finalizados.sum()} ")
```

```python
# Difinir nome do time a ser trabalhado

time = 'SE Palmeiras'

# Criar condição

condicao_palmeiras = (
    (df_brasileirao_2025['mandante'] == time) |
    (df_brasileirao_2025['visitante'] == time)
    )

# Filtrar dados da SE Palmeiras e criar uma copia do DataFrame

df_palmeiras = df_brasileirao_2025[condicao_palmeiras].copy()

# Mostrando primeiras linhas do DataFrame
df_palmeiras.head()
```

### 4.2. Identificação do Local do Jogo e Resultados
Criação de colunas para identificar se o jogo foi em casa ou fora, e para determinar o resultado da partida (Vitória, Empate, Derrota) do ponto de vista do Palmeiras.

```python
# Criar coluna 'local_jogo'
df_palmeiras['local_jogo'] = df_palmeiras['mandante'].apply(lambda x: 'Casa' if x == 'SE Palmeiras' else 'Fora')

# Validação

print("Total de jogos:", df_palmeiras.shape[0])
print(df_palmeiras['local_jogo'].value_counts())
```

```python
df_palmeiras.head()
```

### 4.3. Cálculo de Gols e Pontos
Cálculo dos gols marcados e sofridos pelo Palmeiras em cada partida, e a pontuação obtida (3 para vitória, 1 para empate, 0 para derrota).

```python
# Gols Pró e Gols Contra Ajustados
# Criar coluna 'gols_pro'
df_palmeiras['gols_pro'] = np.where(df_palmeiras['local_jogo'] == 'Casa', df_palmeiras['gols_mandante'], df_palmeiras['gols_visitante'])

# Criar coluna 'gols_contra'
df_palmeiras['gols_contra'] = np.where(df_palmeiras['local_jogo'] == 'Fora', df_palmeiras['gols_mandante'], df_palmeiras['gols_visitante'])
```

```python
# Criando coluna para resultado final do jogo
conditions = [
    df_palmeiras['gols_pro'] > df_palmeiras['gols_contra'],
    df_palmeiras['gols_pro'] == df_palmeiras['gols_contra']
]
choices = ["Vitoria", "Empate"]
df_palmeiras['resultado'] = np.select(conditions, choices, default="Derrota")
```

```python
# Criando coluna para pontos conquistados por partida
condicoes = [
    df_palmeiras['resultado'] == 'Vitoria',
    df_palmeiras['resultado'] == 'Empate'
]
pontuacao =  [3, 1]
df_palmeiras['pontos'] = np.select(condicoes, pontuacao, default = 0)
```

```python
# Criando coluna de saldo de gols

df_palmeiras ['saldo_de_gols'] = df_palmeiras['gols_pro'] - df_palmeiras['gols_contra']
```

### 5.2. Análise de Resultados (Vitórias, Empates, Derrotas)
Contagem e visualização dos resultados das partidas do Palmeiras.

```python
# Totais Gerais
total_vitorias = df_palmeiras['resultado'].value_counts().get('Vitoria', 0)
total_empates = df_palmeiras['resultado'].value_counts().get('Empate', 0)
total_derrotas = df_palmeiras['resultado'].value_counts().get('Derrota', 0)
total_pontos = df_palmeiras['pontos'].sum()
total_gols_pro = df_palmeiras['gols_pro'].sum()
total_gols_contra = df_palmeiras['gols_contra'].sum()
saldo_de_gol = total_gols_pro - total_gols_contra

print(f"Total de Vitórias: {total_vitorias}")
print(f"Total de Empates: {total_empates}")
print(f"Total de Derrotas: {total_derrotas}")
print(f"Total de Pontos: {total_pontos}")
print(f"Total de Gols a Favor: {total_gols_pro}")
print(f"Total de Gols Contra: {total_gols_contra}")
print(f"Saldo de Gols: {saldo_de_gol}")
```

```python
# Calculo de numero de jogos ou numero de rodadas no campeonato
numero_jogos = df_palmeiras.shape[0]

# Calculo de pontos maximos que um time pode fazer
pontos_maximos = numero_jogos * 3

# Calculo de pontos perdidos pelo o palmeiras
pontos_perdidos = pontos_maximos - total_pontos

# Imprime os resultados
print(f"Número de rodadas: {numero_jogos}")
print(f"Total de pontos máximos possiveis no campeonato: {pontos_maximos}")
print(f"Total de pontos conquistados pelo Palmeiras no Campeonato Brasileiro de 2025: {total_pontos}")
print(f"Pontos perdidos pelo Palmeiras no Campeonato Brasileiro de 2025: {pontos_perdidos}")
```

### 5.3. Desempenho em Casa vs. Fora
Comparação da performance do Palmeiras quando joga como mandante e como visitante.

```python
# Desempenho Casa x Fora
resumo_casa_fora = (
    df_palmeiras
        .groupby('local_jogo')
        .agg(
            pontos_total=('pontos', 'sum'),
            gols_pro_total=('gols_pro', 'sum'),
            gols_contra_total=('gols_contra', 'sum'),
            vitorias=('resultado', lambda x: (x == 'Vitoria').sum()),
            empates=('resultado', lambda x: (x == 'Empate').sum()),
            derrotas=('resultado', lambda x: (x == 'Derrota').sum())
        )
        .reset_index()
)

resumo_casa_fora
```

```python
# Métricas percentuais

# Aproveitamento de pontos
aproveitamento_percentual = total_pontos / pontos_maximos * 100

# Média de gols feitos
media_gols_pro = total_gols_pro / numero_jogos

# Média gols contra
media_gols_contra = total_gols_contra / numero_jogos

print(f"Percentual de pontos conquistados: {aproveitamento_percentual:.2f}%")
print(f"Média de gols feitos por jogo: {media_gols_pro:.2f}")
print(f"Média de gols sofridos por jogo: {media_gols_contra:.2f}")
```

## 6. Análise de Vítimas e Carrascos

### 6.1. Maiores Vítimas do Palmeiras
Identificação dos times que o Palmeiras mais venceu no Campeonato Brasileiro de 2025.

```python
# Identificando as vítimas (times que o Palmeiras mais venceu)
vitorias_mandante = df_palmeiras[(df_palmeiras['mandante'] == time) & (df_palmeiras['resultado'] == 'Vitoria')]['visitante'].value_counts()
vitorias_visitante = df_palmeiras[(df_palmeiras['visitante'] == time) & (df_palmeiras['resultado'] == 'Vitoria')]['mandante'].value_counts()

maiores_vitimas = (vitorias_mandante.add(vitorias_visitante, fill_value=0)).sort_values(ascending=False)
print("Maiores Vítimas do Palmeiras:")
print(maiores_vitimas.head())
```

### 6.2. Maiores Carrascos do Palmeiras
Identificação dos times que mais venceram o Palmeiras no Campeonato Brasileiro de 2025.

```python
# Identificando os carrascos (times que mais venceram o Palmeiras)
derrotas_mandante = df_palmeiras[(df_palmeiras['mandante'] == time) & (df_palmeiras['resultado'] == 'Derrota')]['visitante'].value_counts()
derrotas_visitante = df_palmeiras[(df_palmeiras['visitante'] == time) & (df_palmeiras['resultado'] == 'Derrota')]['mandante'].value_counts()

maiores_carrascos = (derrotas_mandante.add(derrotas_visitante, fill_value=0)).sort_values(ascending=False)
print("""
Maiores Carrascos do Palmeiras:
""")
print(maiores_carrascos.head())
```
